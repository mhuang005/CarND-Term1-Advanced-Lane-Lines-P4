# Advanced Lane Finding

The goal of this project is to build a pipeline used for identifying lane lines in a video.
This pipelin includes the following steps:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.



## Camera Calibration

* Compute the camera calibration matrix and distortion coefficients using chessboard images in the directory "camera_cal".
* Un-distort one of images in the directory "camera_cal" to check if the camera calibration works correctly.

```python
import numpy as np
import cv2
import glob
import pickle
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from copy import deepcopy
import imageio
imageio.plugins.ffmpeg.download()
from moviepy.editor import VideoFileClip
from IPython.display import HTML
%matplotlib qt
%matplotlib inline

# prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(8,5,0)
objp = np.zeros((6*9,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

# Arrays to store object points and image points from all the images.
objpoints = [] # 3d points in real world space
imgpoints = [] # 2d points in image plane.

# Make a list of calibration images
images = glob.glob('./camera_cal/calibration*.jpg')

# Step through the list and search for chessboard corners
for fname in images:
    img = cv2.imread(fname)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # Find the chessboard corners
    ret, corners = cv2.findChessboardCorners(gray, (9,6),None)

    # If found, add object points, image points
    if ret == True:
        objpoints.append(objp)
        imgpoints.append(corners)

        # Draw and display the corners
        img = cv2.drawChessboardCorners(img, (9,6), corners, ret)
        cv2.imshow('img',img)
        cv2.waitKey(500)

cv2.destroyAllWindows()
```

## Helper Functions for  Thresholded Binary Image Generation

The functnions we defined here are used for 

* distortion correction
* color channel and gradient channel threshold
* perspective transform

```python
def undistort(img, mtx, dist):
    return cv2.undistort(img, mtx, dist, None, mtx)

# Define a function that separates 3 channels of a colorful RGB image
# and then applies a threshold to one of 3 channels.
def hls_thresh(img, channel='S', thresh=(0, 255)):
    # Convert to HLS color space 
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    
    hls_binary = np.zeros_like(hls[:,:,0])
    if (channel == 'H'):
        hls_binary[(hls[:,:,0] >= thresh[0]) & (hls[:,:,0] <= thresh[1])] = 1
    elif (channel == 'L'):
        hls_binary[(hls[:,:,1] >= thresh[0]) & (hls[:,:,1] <= thresh[1])] = 1
    elif (channel == 'S'):
        hls_binary[(hls[:,:,2] >= thresh[0]) & (hls[:,:,2] <= thresh[1])] = 1
        
    return hls_binary    
    
# Define a function that applies Sobel x or y, 
# then applies a threshold.
def abs_sobel_thresh(img, orient='x', thresh=(0, 255), sobel_kernel=3):
    # Convert to grayscale if necessary
    if (len(img.shape) > 2):
        img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        
    # Take the derivative in x or y and scale to 8 bit
    dx, dy = 0, 0
    if orient == 'x': dx = 1
    if orient == 'y': dy = 1
    abs_sobel = np.abs(cv2.Sobel(img, cv2.CV_64F, dx, dy, ksize=sobel_kernel))
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    
    # Create a binary mask where the gradient thresholds are met
    grad_binary = np.zeros_like(scaled_sobel)
    grad_binary[(scaled_sobel >= thresh[0]) & (scaled_sobel <= thresh[1])] = 1
    
    return grad_binary


# Define a function that applies color and Sobel gradient thresholds to
# generate a binary image
def binary(img, s_thresh=(100, 255), l_thresh=(40, 255), sx_thresh=(20, 100), sy_thresh=(30, 100), sobel_kernel=3):
    gradx_binary = abs_sobel_thresh(img, 'x', sx_thresh)
    grady_binary = abs_sobel_thresh(img, 'y', sy_thresh)
    s_binary = hls_thresh(img, 'S', s_thresh)
    l_binary = hls_thresh(img, 'L', l_thresh)
    combined_binary = np.zeros_like(s_binary)
    combined_binary[((gradx_binary == 1) & (grady_binary == 1) ) | ((s_binary == 1) & (l_binary == 1) )] = 1
    
    return combined_binary

# Define a perspective transfom funciton
def warper(img, src, dist):
    img_size = (img.shape[1], img.shape[0]) 
    # Calculate the perspective transform matrix
    M = cv2.getPerspectiveTransform(src, dst)
    Minv = cv2.getPerspectiveTransform(dst, src)
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    
    return warped, M, Minv
    
```

### Helper Funcitons for Lane Line Finding

The functions are used for

* identifying lane line pixels and fitting to polynomials
* calculating the radius of curvature and vehicle position
* warping lane lines back to the original image and displaying the output 

```python
# Introduce a dictionary which tracks the lane lines in a video frame by frame
line = {'detected': False, #  was the line detected in the last iteration?
        'recent_fit': [], # polynomial coefficients of the last n fits of the line
        'avg_fit':  None, # polynomial coefficients averaged over the last n iterations
        'current_fit':  [np.array([False])], # polynomial coefficients for the most recent fit        
        'allx':  None, # x values for detected line pixels
        'ally':  None # y values for detected line pixels
}

# Find lane line pixels using histogram and then fit to quadratic polynomials
def detect_lines(binary_warped):
    # Assuming you have created a warped binary image called "binary_warped"
    # Take a histogram of the bottom half of the image
    histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:], axis=0)
    
    # Create an output image to draw on and  visualize the result
    out_img = np.dstack((binary_warped, binary_warped, binary_warped))*255
    
    # Find the peak of the left and right halves of the histogram
    # These will be the starting point for the left and right lines
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint

    # Choose the number of sliding windows and set height of windows
    nwindows = 9
    window_height = np.int(binary_warped.shape[0]//nwindows)
    
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    
    # Current positions to be updated for each window
    leftx_current = leftx_base
    rightx_current = rightx_base
    
    # Set the width of the windows +/- margin
    margin = 100

    # Set minimum number of pixels found to recenter window
    minpix = 50

    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []

    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        win_xleft_low = leftx_current - margin
        win_xleft_high = leftx_current + margin
        win_xright_low = rightx_current - margin
        win_xright_high = rightx_current + margin
        
        # Draw the windows on the visualization image
        cv2.rectangle(out_img,(win_xleft_low,win_y_low),(win_xleft_high,win_y_high),(0,255,0), 2) 
        cv2.rectangle(out_img,(win_xright_low,win_y_low),(win_xright_high,win_y_high),(0,255,0), 2) 
   
        # Identify the nonzero pixels in x and y within the window
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
                          (nonzerox >= win_xleft_low) &  (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
                           (nonzerox >= win_xright_low) &  (nonzerox < win_xright_high)).nonzero()[0]
    
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        
        # If you found > minpix pixels, recenter next window on their mean position
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))

    # Concatenate the arrays of indices
    left_lane_inds = np.concatenate(left_lane_inds)
    right_lane_inds = np.concatenate(right_lane_inds)

    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds] 

    # Fit a second order polynomial to each
    left_fit = np.polyfit(lefty, leftx, 2)
    right_fit = np.polyfit(righty, rightx, 2)
    
    return left_fit, right_fit, leftx, lefty, rightx, righty

def update_lines(lane, current_fit, allx, ally, detected=True):
    # Update line once found the polynomial to fit the line
    lane['current_fit'] = current_fit
    lane['recent_fit'].append(current_fit)
    lane['recent_fit'] = lane['recent_fit'][-5:]
    lane['avg_fit'] = np.mean(lane['recent_fit'], axis=0)
    lane['detected'] = detected
    lane['allx'] = allx
    lane['ally'] = ally
    
# Find the lane line pixels in a new (video) frame based on 
# quadratic polynomials obtaine from previous frames (i.e. starting point)
def generate_lines(warped, left, right):
    # Find polnomials to fit the left and right lines, respectively,
    # and then update the dictinories 'left' and 'right'
    if not (left['detected'] and right['detected']):
        left_fit, right_fit, leftx, lefty, rightx, righty = detect_lines(warped)
        
        # Update left and right
        update_lines(left, left_fit, leftx, lefty)
        update_lines(right, right_fit, rightx, righty)

    else:
        # Use previous lines for a new search of current lines
        #left_fit = left['current_fit']
        #right_fit = right['current_fit']
        left_fit = left['avg_fit']
        right_fit = right['avg_fit']
        
        nonzero = warped.nonzero()
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])
        margin = 100
        left_lane_inds = ((nonzerox > (left_fit[0]*(nonzeroy**2) + left_fit[1]*nonzeroy + left_fit[2] - margin)) & 
                          (nonzerox < (left_fit[0]*(nonzeroy**2) + left_fit[1]*nonzeroy + left_fit[2] + margin))) 

        right_lane_inds = ((nonzerox > (right_fit[0]*(nonzeroy**2) + right_fit[1]*nonzeroy + right_fit[2] - margin)) & 
                           (nonzerox < (right_fit[0]*(nonzeroy**2) + right_fit[1]*nonzeroy + right_fit[2] + margin)))  

        # Again, extract left and right line pixel positions
        leftx = nonzerox[left_lane_inds]
        lefty = nonzeroy[left_lane_inds] 
        rightx = nonzerox[right_lane_inds]
        righty = nonzeroy[right_lane_inds]
    
        # Fit a second order polynomial to each
        left_fit = np.polyfit(lefty, leftx, 2)
        right_fit = np.polyfit(righty, rightx, 2)
        
        # Update left and right
        update_lines(left, left_fit, leftx, lefty)
        update_lines(right, right_fit, rightx, righty)
        
        
def compute_radius_and_deviation(warped, left, right):
    leftx = left['allx']
    lefty = left['ally']
    rightx = right['allx']
    righty = right['ally']
    
    # Define conversions in x and y from pixels space to meters
    ym_per_pix = 30/720   # meters per pixel in y dimension
    xm_per_pix = 3.7/700  # meters per pixel in x dimension
    
    # Define y-value where we want radius of curvature
    (x_val, y_eval) = (warped.shape[1], warped.shape[0])
 
    # Fit new polynomials to x,y in world space
    left_fit_cr = np.polyfit(lefty*ym_per_pix, leftx*xm_per_pix, 2)
    right_fit_cr = np.polyfit(righty*ym_per_pix, rightx*xm_per_pix, 2)

    # Calculate the new radius of curvature
    left_radius = ((1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    right_radius = ((1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
    radius = 0.5*(left_radius+right_radius)
    
    # Calculate the deviation of car from the center
    center = 0.5*(np.polyval(left_fit_cr, 0.5*y_eval*ym_per_pix) + np.polyval(right_fit_cr, 0.5*y_eval*ym_per_pix))
    deviation = 0.5*x_val*xm_per_pix - center
    
    return radius, deviation

# Warp the lane boundaries back to the original image and then visulizing the output 
def plot_lines(undist, warped, Minv, left, right):
    # Warp the lane boundaries back to the original image
    # and then visulizing the output 
    left_fit = left['avg_fit']
    right_fit = right['avg_fit']
    
    # Create an image to plot lines on
    color_warped = np.zeros_like(undist).astype(np.uint8)
    
    # Generate x and y values for plotting
    ploty = np.linspace(0, warped.shape[0]-1, warped.shape[0] )
    left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    
    # Generate a polygon and recast the x and y points into usable format for cv2.fillPoly()
    left_line_pts = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    right_line_pts = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    line_pts = np.hstack((left_line_pts, right_line_pts))
    
    cv2.fillPoly(color_warped, np.int_([line_pts]), (0,255, 0))
    
    # Warp back to undistorted image
    warped_back = cv2.warpPerspective(color_warped, Minv, (warped.shape[1], warped.shape[0]), flags=cv2.INTER_LINEAR)
    
    # Combine two images
    final_img = cv2.addWeighted(undist, 1, warped_back, 0.3, 0)
    
    # Calculate the radius and deviation (of car from the center)
    rad, dev = compute_radius_and_deviation(warped, left, right)
    
    # Add descriptions of radius and deviation to the final image
    rad_text = "Radius of the curvature: {} m".format(round(rad, 3))
    if dev >= 0:
        dev_text = "Right of the center: {} m".format(round(dev, 3))
    else:
        dev_text = "Left of the center: {} m".format(round(-dev, 3))

    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(final_img, rad_text, (20,50), font, 1,(255,255,255),3,cv2.LINE_AA)
    cv2.putText(final_img, dev_text,(20,100), font, 1,(255,255,255),3,cv2.LINE_AA)
    
    return final_img
  
def pipeline(img):
    global mtx, dist, src, dst 
    global left, right

    # Undistort the image
    undist = undistort(img, mtx, dist)
    
    # Apply the color and gradient thresholds
    combined_binary = binary(undist)
    
    # Apply a perspective transform to rectify a binary image
    warped, M, Minv = warper(combined_binary, src, dst)
    
    # Detect lane pixels and fit to find the land boundaries
    generate_lines(warped, left, right)
    
    # Warp the detected lane boundaries back to the original image
    # and display the boundaries on the image with estimation of
    # lane curvature and vehicle position
    final_img = plot_lines(undist, warped, Minv, left, right)
      
    return final_img          
```
